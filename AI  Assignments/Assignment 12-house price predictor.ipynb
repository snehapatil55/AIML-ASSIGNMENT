## Assignment Name: House Price Predictor

## Description
Train a Linear Regression model, predict house prices, and test with new input.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Create dataset
data = {
    "Area": [500, 800, 1000, 1200, 1500, 1800, 2000],
    "Price": [100000, 150000, 200000, 240000, 300000, 360000, 400000]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)

# Features and Labels
X = df[["Area"]]
y = df["Price"]

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict new value
new_data = pd.DataFrame([[1700]], columns=["Area"])
prediction = model.predict(new_data)

print("\nPredicted Price for 1700 sq ft:", round(prediction[0], 2))
```
