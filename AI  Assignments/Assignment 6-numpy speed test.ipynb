## Assignment Name: NumPy Speed Test

## Description
Compare Python lists vs NumPy arrays with 1 million numbers, measure execution time.

```python
import time
import numpy as np

# Create 1 million numbers
size = 1000000

# Python List
start = time.time()
py_list = list(range(size))
py_result = [x * 2 for x in py_list]
end = time.time()

list_time = end - start
print("Python List Time:", list_time)


# NumPy Array
start = time.time()
np_array = np.arange(size)
np_result = np_array * 2
end = time.time()

numpy_time = end - start
print("NumPy Array Time:", numpy_time)
```
