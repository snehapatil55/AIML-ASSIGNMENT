# Assignment (10/03/2026)

## Assignment Name: Spam Classifier Thinking

## Description : Design a spam detection system: features, data needed, possible mistakes.

## 1. Overview

A spam detection system classifies messages as **Spam** or **Not Spam** using machine learning.

---

## 2. System Flow (Design)

Input Message → Preprocessing → Feature Extraction → ML Model → Output (Spam / Not Spam)

---

## 3. Data Required

- Dataset of emails/SMS  
- Labeled as:
  - Spam  
  - Not Spam  

---

## 4. Preprocessing

- Convert text to lowercase  
- Remove punctuation and stopwords  
- Tokenization (split words)

---

## 5. Features

- Spam keywords (free, win, offer)  
- Number of links  
- Message length  
- Special characters (!, $$$)  
- Sender details  

---

## 6. Model Used

- Naive Bayes (commonly used for text classification)  
- Logistic Regression (alternative)

---

## 7. Output

- Spam  
- Not Spam  

---

## 8. Possible Errors

- False Positive: Normal message marked as spam  
- False Negative: Spam message not detected  

---

## 9. Conclusion

The system learns patterns from data and helps filter unwanted messages automatically.
