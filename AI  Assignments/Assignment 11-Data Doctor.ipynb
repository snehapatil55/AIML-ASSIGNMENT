### Clean a dataset by handling missing values, removing duplicates, standardizing text, and explain why cleaning matters.

```python
import pandas as pd

# Load dataset (using online dataset)
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")

print("Original Data:")
print(df.head())

# Introduce some missing values (for demonstration)
df.loc[0, 'sepal_length'] = None
df.loc[1, 'species'] = None

# Handle missing values
df.fillna(method='ffill', inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Standardize text (convert to lowercase)
df['species'] = df['species'].str.lower()

print("\nCleaned Data:")
print(df.head())
```
