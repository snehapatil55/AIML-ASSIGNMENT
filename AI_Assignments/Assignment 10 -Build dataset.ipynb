## Description
### Create a dataset (study hours vs marks), identify features and labels, and predict the relationship.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Create dataset
data = {
    "Study Hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Marks": [30, 35, 50, 55, 65, 70, 80, 90]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)

# Features and Labels
X = df[["Study Hours"]]   # Feature
y = df["Marks"]           # Label

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict marks for 9 hours (FIXED - no warning)
new_data = pd.DataFrame([[9]], columns=["Study Hours"])
prediction = model.predict(new_data)

print("\nPredicted Marks for 9 hours study:", round(prediction[0], 2))
```
