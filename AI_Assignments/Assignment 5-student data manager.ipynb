# Assignment: Student Data Manager

## Description
Store data for 5 students using dictionaries, print topper, class average, and assign grades.

## Python Program

```python
def assign_grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 75:
        return 'B'
    elif marks >= 60:
        return 'C'
    elif marks >= 50:
        return 'D'
    else:
        return 'F'


def student_data_manager():
    students = {}

    # Input data for 5 students
    for i in range(5):
        name = input("Enter name of student {}: ".format(i+1))
        
        # Input validation (important improvement)
        while True:
            try:
                marks = int(input("Enter marks of {}: ".format(name)))
                if 0 <= marks <= 100:
                    break
                else:
                    print("Enter marks between 0 and 100.")
            except:
                print("Invalid input. Enter a number.")

        students[name] = marks

    # Find topper
    topper = max(students, key=students.get)

    # Calculate average
    average = sum(students.values()) / len(students)

    print("\n--- Student Details ---")
    for name, marks in students.items():
        grade = assign_grade(marks)
        print("{}: Marks = {}, Grade = {}".format(name, marks, grade))

    print("\nTopper: {} - {}".format(topper, students[topper]))
    print("Class Average: {:.2f}".format(average))


# Run the function
student_data_manager()
```
