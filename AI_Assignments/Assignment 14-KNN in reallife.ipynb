## KNN in Real Life (Netflix Example)

K-Nearest Neighbors (KNN) recommends movies based on similarity between users.

- It finds users with similar watching patterns  
- Suggests movies liked by similar users  
- "K" represents number of nearest neighbors  

Example:
If User A and User B have similar taste, movies watched by User B will be recommended to User A.

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity


data = np.array([
    [5, 4, 0],   # User A
    [4, 5, 0],   # User B
    [0, 0, 5]    # User C
])


similarity = cosine_similarity(data)

print("Similarity Matrix:\n", similarity)
most_similar = np.argmax(similarity[0][1:]) + 1
print("\nUser most similar to User A: User", chr(65 + most_similar))
```
