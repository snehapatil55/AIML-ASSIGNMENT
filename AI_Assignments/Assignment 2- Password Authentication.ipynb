## Create a strong code for password authentication using python. 



```python
def check_password(password):
    
    if len(password) < 8:
        print("Password too short")
        return False
    
    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False
    
    for ch in password:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True
        elif ch in "@#$%&*!":
            has_special = True
    
    if not has_upper:
        print("Need uppercase letter")
    if not has_lower:
        print("Need lowercase letter")
    if not has_digit:
        print("Need number")
    if not has_special:
        print("Need special character")
    
    return has_upper and has_lower and has_digit and has_special


correct_password = "Admin@123"

attempts = 3

while attempts > 0:
    user_pass = input("Enter password: ")
    
    if check_password(user_pass):
        if user_pass == correct_password:
            print("Login successful")
            break
        else:
            print("Wrong password")
    else:
        print("Weak password")
    
    attempts -= 1
    print("Attempts left:", attempts)

if attempts == 0:
    print("Account locked")
    
```
