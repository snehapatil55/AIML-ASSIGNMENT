## Assignment Name: Build a Text Cleaner
### Description

#### Write code to clean text by removing punctuation, converting to lowercase, and removing stopwords.

```python
import string
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# Sample messy sentences
sentences = [
    "OMG!!! this movie is sooo gud ",
    "lol bro u coming tmrw??",
    "I cant beleive this happnd!!!"
]

def clean_text(text):
    # Lowercase
    text = text.lower()
    
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Remove stopwords
    words = text.split()
    words = [word for word in words if word not in ENGLISH_STOP_WORDS]
    
    return " ".join(words)

# Apply cleaning
for s in sentences:
    print("Original:", s)
    print("Cleaned:", clean_text(s))
    print()
```
