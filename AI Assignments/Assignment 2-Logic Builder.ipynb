## Logic Builder

```python
def fizz_buzz():
    fizz = 0
    buzz = 0
    fizzbuzz = 0

    for i in range(1, 51):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz", end=" ")
            fizzbuzz += 1
        elif i % 3 == 0:
            print("Fizz", end=" ")
            fizz += 1
        elif i % 5 == 0:
            print("Buzz", end=" ")
            buzz += 1
        else:
            print(i, end=" ")

    print("\n\nCount:")
    print("Fizz:", fizz)
    print("Buzz:", buzz)
    print("FizzBuzz:", fizzbuzz)



fizz_buzz()
```
