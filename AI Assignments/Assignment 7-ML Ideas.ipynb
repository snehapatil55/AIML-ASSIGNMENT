## Assignment Name: ML Idea Generator
# Description:
### Suggest ML problems in college, healthcare, and shopping, and describe input → output.

# ML Ideas

## 1. College – Student Performance Prediction
- **Problem:** Predict student marks or grades  
- **Input:** Attendance, previous marks, study hours  
- **Output:** Predicted grade (A/B/C/D)  

## 2. College – Attendance Monitoring (Face Recognition)
- **Problem:** Automatically mark attendance  
- **Input:** Student face images (camera input)  
- **Output:** Attendance status (Present/Absent)  

## 3. Healthcare – Disease Prediction
- **Problem:** Predict diseases based on symptoms  
- **Input:** Symptoms (fever, cough, BP, etc.)  
- **Output:** Possible disease (e.g., flu, diabetes)  

## 4. Healthcare – Medical Image Analysis
- **Problem:** Detect abnormalities in X-rays or scans  
- **Input:** Medical images (X-ray, MRI)  
- **Output:** Detection result (Normal / Abnormal)  

## 5. Shopping – Product Recommendation System
- **Problem:** Suggest products to users  
- **Input:** User browsing history, purchases  
- **Output:** Recommended products  

## 6. Shopping – Price Prediction
- **Problem:** Predict product price trends  
- **Input:** Past prices, demand, season  
- **Output:** Future price estimate
