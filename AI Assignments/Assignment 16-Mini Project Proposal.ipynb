# Assignment (19/03/2026)
## Assignment Name: Mini Project Proposal

### Project Title: Student Performance Prediction using Machine Learning

## Problem Statement
...Many students struggle to track and improve their academic performance.
This project aims to predict student marks based on study habits and past performance, helping identify weak areas early.

## Dataset
...Type: Structured dataset (CSV)
Data includes:
Study hours
Attendance percentage
Previous marks
Dataset can be:
Self-created OR
Taken from online sources

## Algorithm Used
...Linear Regression
Reason:
Simple and suitable for predicting continuous values (marks)
Shows relationship between input features and output

## Methodology
...Collect dataset
Preprocess data (cleaning, handling missing values)
Train model using Linear Regression
Predict marks for new student data

## Expected Output
...Predicted student marks
Graph showing relationship between study hours and marks
Insights on how performance improves with input factors

## Conclusion
...This project helps in understanding how different factors affect student performance and can be useful for early academic improvement.
