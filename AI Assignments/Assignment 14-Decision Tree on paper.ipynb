# Assignment (12/03/2026)

## Assignment Name : Decision Tree on Paper
### Description : Draw a decision tree predicting whether you should play outside.

## Decision Tree

We predict whether to play outside based on weather conditions like Outlook, Temperature, and Humidity.

## Decision Tree Diagram

           Outlook
         /    |     \
      Sunny Overcast Rain
       /         |       \
Humidity       Yes      Wind
   /   \                  /   \
High Normal            Strong Weak
 No     Yes             No      Yes

## 

- If outlook is Overcast → Play  
- If Sunny → check humidity  
  - High → Do not play  
  - Normal → Play  
- If Rain → check wind  
  - Strong → Do not play  
  - Weak → Play
